```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
sns.lineplot(data = tips,
             x= 'total_bill',
             y='tip')
```

```python
tipgraph=sns.lineplot(data=tips,
                      x='total_bill',
                      y='tip')

tipgraph.set(title='Tips vs. Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
fig =plt.subplots(figsize=(10,3))

tipgrap1 = sns.lineplot(data = tips,
                        x= 'total_bill',
                        y= 'tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth =3,
                        marker = 'o',
                        markerfacecolor = 'indigo')
tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill',
             ylabel = 'Tip Amount ($)')
```

```python
tipgrap1 = sns.scatterplot(data = tips,
                        x= 'total_bill',
                        y= 'tip',
                        style = 'time',
                        hue = 'day',
                        palette = ['purple','hotpink','deepskeyblue','yellowgreen'])
tipgraph.set(title = 'Tips vs. Total Bill',
             xlabel = 'Total Bill',
             ylabel = 'Tip Amount ($)')
```

```python
sns.pairplot(data=tips,
             hue = 'day')
```

```python
daysplot = sns.relplot(data=tips,
                       x='total_bill',
                       y='tip',
                       hue='sex',
                       col='day',
                       kind = 'scatter',
                       col_wrap =2)

daysplot
```

```python
import plotly.graph_objects as go

x1=np.array(['Luffy','Zoro','Nami','Usopp','Sanji'])
y1= np.array([110,180,240,99,220])
```

```python
graph = go.Figure()

graph.add_trace(go.Bar(x=x1,y=y1))

graph.update_layout(title = "Pirate Scores",
                    xaxis_title = "pirates",
                    yaxis_title="Score")
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(
    x=x1,
    y=y1,
    mode='markers',
    marker=dict(
        size=15,
        color='hotpink',
        opacity=1,
        line=dict(width=5,
                  color='purple')
    )
))
graph.update_layout(title='Interactive Pirate Plot',
                    xaxis_title="Pirates",
                    yaxis_title="Scores",
                    width=500,
                    height=500)
```
